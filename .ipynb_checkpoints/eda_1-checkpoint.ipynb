## Configurar Spark

```python
import os
import sys
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, log1p

# --- CONFIGURACIÓN SPARK (Tu bloque) ---
os.environ['SPARK_HOME'] = "/home/hadoop/spark"
sys.path.insert(0, "/home/hadoop/spark/python")
sys.path.insert(0, "/home/hadoop/spark/python/lib/py4j-0.10.9.7-src.zip") 
sys.path.insert(0, "/home/hadoop/spark/python/lib/pyspark.zip")

spark = SparkSession.builder \
    .appName("EDA_Ciberseguridad") \
    .master("yarn") \
    .config("spark.executor.memory", "4g") \
    .config("spark.executor.cores", "2") \
    .config("spark.yarn.executor.memoryOverhead", "1024m") \
    .getOrCreate()

print("✅ Spark iniciado.")
```

```python
df = spark.read.parquet("/trafico_clean")
df_sample_cache = df.sample(fraction=0.01, seed=42).limit(100000).cache()
print("✅ Datos cargados y muestra en caché.")
```

## Distribución de variable onjetivo

```python
print("\nDistribución de clases")

distribucion = df.groupBy("Label").count().orderBy("count", ascending=False).toPandas()

total = distribucion["count"].sum()
distribucion["porcentaje"] = (distribucion["count"] / total) * 100

print(distribucion)

# Gráfica
plt.figure(figsize=(14, 7))
sns.barplot(x="count", y="Label", data=distribucion, palette="plasma")
plt.title("Distribución de clases ")
plt.xlabel("Cantidad de Registros")
plt.xscale("log") 
plt.grid(axis="x", alpha=0.3)
plt.show()
```

## Correlación y redundancia

```python
print("\nCorrelación")

cols_num = [c for c in df.columns if c != "Label"]

pdf_corr = df_sample_cache.select(cols_num).toPandas()

corr_matrix = pdf_corr.corr()

plt.figure(figsize=(20, 16))
sns.heatmap(corr_matrix, cmap="coolwarm", vmin=-1, vmax=1, square=True)
plt.title("Mapa de Correlación de Características")
plt.show()

print("🔍 Buscando pares altamente correlacionados (> 0.95):")
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column].abs() > 0.95)]
print(f"Variables redundantes detectadas ({len(to_drop)}): {to_drop[:10]} ...")
```

```python

print("Generando tabla de redundancias (> 0.95)...")

pares_corr = upper.stack()#aplanar

pares_altos = pares_corr[pares_corr.abs() > 0.95]

df_redundancia = pd.DataFrame(pares_altos).reset_index()
df_redundancia.columns = ['Variable A', 'Variable B', 'Correlación']

df_redundancia['Abs_Corr'] = df_redundancia['Correlación'].abs()
df_redundancia = df_redundancia.sort_values(by='Abs_Corr', ascending=False).drop(columns=['Abs_Corr'])


print(f"\nSe encontraron {len(df_redundancia)} pares de variables redundantes:")
print("-" * 80)
print(df_redundancia.head(30))
print("-" * 80)

to_drop = df_redundancia['Variable B'].unique().tolist()
print(f"Posibles a eliminar: {len(to_drop)} columnas redundantes para optimizar el modelo:")
print(to_drop)
```

```python
print("\nSeparabilidad ")

pdf_plot = df_sample_cache.select("Label", "Flow Duration", "Tot Fwd Pkts", "Pkt Len Mean").toPandas()
pdf_plot["Tipo"] = pdf_plot["Label"].apply(lambda x: "Benign" if x == "Benign" else "Attack")

features_clave = ["Flow Duration", "Tot Fwd Pkts", "Pkt Len Mean"]

plt.figure(figsize=(15, 5))
for i, col_name in enumerate(features_clave):
    plt.subplot(1, 3, i+1)
    sns.boxplot(x="Tipo", y=col_name, data=pdf_plot, palette="Set2")
    plt.yscale("log") 
    plt.title(f"Distribución de {col_name}")

plt.tight_layout()
plt.show()
```

```python
print("\nDDoS vs Infiltration")

clases_interes = ["Benign", "DDOS attack-HOIC", "Infilteration"]
df_spec = df.filter(col("Label").isin(clases_interes))

# Tomamos muestra de estas clases específicas
pdf_spec = df_spec.select("Label", "Flow Byts/s", "Flow IAT Mean").limit(50000).toPandas()

# Gráfico de dispersión (Scatter plot) para ver patrones
plt.figure(figsize=(12, 6))
sns.scatterplot(data=pdf_spec, x="Flow IAT Mean", y="Flow Byts/s", hue="Label", alpha=0.6)
plt.yscale("log")
plt.xscale("log")
plt.title("Patrones: Tasa de Transferencia vs Tiempos (DDoS vs Infiltration)")
plt.show()
```

```python
print("\n📈 Histogramas")

# Usamos la muestra cacheada
pdf_hist = df_sample_cache.toPandas()
pdf_hist["Clase"] = pdf_hist["Label"].apply(lambda x: "Normal" if x == "Benign" else "Ataque")

# Variables sospechosas de ser importantes (puedes cambiar estas)
vars_importantes = ["Fwd Seg Size Min", "Init Bwd Win Byts", "Ack Flag Cnt"] 

plt.figure(figsize=(15, 5))
for i, var in enumerate(vars_importantes):
    plt.subplot(1, 3, i+1)
    # Histograma superpuesto
    sns.histplot(data=pdf_hist, x=var, hue="Clase", element="step", stat="density", common_norm=False, log_scale=True)
    plt.title(f"Separación en {var}")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from pyspark.ml import PipelineModel

print("📂 Cargando el modelo guardado...")
ruta_modelo = "/modelos/rf_all_features_v1"

# 1. Cargar el Pipeline completo (que incluye los 4 stages que viste)
model_cargado = PipelineModel.load(ruta_modelo)

# 2. Identificar los Stages (Etapas)
# Según tu 'ls', el orden es:
# 0: StringIndexer
# 1: VectorAssembler (Aquí están los nombres de las columnas originales)
# 2: StandardScaler
# 3: RandomForestClassifier (Aquí están los valores de importancia)

assembler_stage = model_cargado.stages[1] # VectorAssembler
rf_stage = model_cargado.stages[-1]       # RandomForest (el último)

print("✅ Modelo cargado exitosamente.")

# 3. Extraer los Nombres de las Columnas
# El VectorAssembler guarda qué columnas usó como entrada
nombres_columnas = assembler_stage.getInputCols()

# 4. Extraer los Valores de Importancia
# El RandomForest tiene el atributo 'featureImportances'
importancias = rf_stage.featureImportances.toArray()

# 5. Unir Nombres con Valores y Ordenar
df_importancia = pd.DataFrame({
    'Columna': nombres_columnas,
    'Importancia': importancias
}).sort_values(by='Importancia', ascending=False)

# 6. Mostrar el Top 20
print("\n🏆 TOP 20 VARIABLES MÁS IMPORTANTES:")
print(df_importancia.head(20))

# Opcional: Graficar
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.barplot(x='Importancia', y='Columna', data=df_importancia.head(20), palette='viridis')
plt.title('Importancia de Características (Modelo Recuperado)')
plt.show()
```
